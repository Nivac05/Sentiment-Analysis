```python
import pandas as pd
import json
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Function to get sentiment
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Apply the function to the 'body' column
merged_data_df['TextBlob_score'] = merged_data_df['body'].apply(get_sentiment)
```

```python
merged_data_df.drop(columns=['symbols', 'mentioned_users',
        'likes', 'conversation', 'links', 'reshare_message',
       'reshares', 'owned_symbols','user', 'source'], inplace=True)
```

```python
# TextBlob Sentiment Analysis
def textblob_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'

merged_data_df['TextBlob_Sentiment'] = merged_data_df['body'].apply(textblob_sentiment)

# VADER Sentiment Analysis
vader_analyzer = SentimentIntensityAnalyzer()

def vader_sentiment(text):
    scores = vader_analyzer.polarity_scores(text)
    if scores['compound'] >= 0.05:
        return 'positive'
    elif scores['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

def vader_sentiment1(text):
    scores = vader_analyzer.polarity_scores(text)
    return scores['compound']


merged_data_df['VADER_Sentiment'] = merged_data_df['body'].apply(vader_sentiment)
merged_data_df['VADER_Score'] = merged_data_df['body'].apply(vader_sentiment1)
```

```python
textblob_counts = merged_data_df['TextBlob_Sentiment'].value_counts()
vader_counts = merged_data_df['VADER_Sentiment'].value_counts()

import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val})'
    return my_autopct

plt.subplot(1, 2, 1)
textblob_counts.plot(kind='pie', autopct=make_autopct(textblob_counts), colors=['#ff9999','#66b3ff','#99ff99'], startangle=140)
plt.title('TextBlob Sentiment Distribution')
plt.ylabel('')


plt.subplot(1, 2, 2)
vader_counts.plot(kind='pie', autopct=make_autopct(vader_counts), colors=['#ff9999','#66b3ff','#99ff99'], startangle=140)
plt.title('VADER Sentiment Distribution')
plt.ylabel('')

plt.tight_layout()
plt.show()
```
